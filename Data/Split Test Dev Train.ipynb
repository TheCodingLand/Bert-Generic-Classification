```python
from sklearn.model_selection import train_test_split
import pandas as pd
trainfile = "train.tsv"
trainfile = "dev.tsv"
datafile = "MILLICOM (1)"
```

```python
ml_dataset = pd.read_csv(datafile)
ml_dataset = ml_dataset.groupby('Subcategory').filter(lambda L: len(L) > 10)
#df.groupby('Subcategory').filter(lambda x: len(x) > 10).drop_duplicates(subset='a')
label_map = ml_dataset['Subcategory'].unique()

ml_dataset['text'] = ml_dataset['Subject'] + ' ' + ml_dataset['Description']
ml_dataset['fakelabel'] = '*'
ml_dataset_prepared = ml_dataset[['Subcategory', 'fakelabel','text']]
ml_dataset_prepared.head()
#ml_dataset_prepared['Subject'] = ml_dataset[['Subject']] + ' ' + ml_dataset[['Description']]
```

```python
train, test = train_test_split(ml_dataset_prepared,train_size=0.85,stratify=ml_dataset_prepared['Subcategory'])
```

```python
train.to_csv('train.tsv',header=False,sep='\t')
test.to_csv('dev.tsv',header=False,sep='\t')
```
